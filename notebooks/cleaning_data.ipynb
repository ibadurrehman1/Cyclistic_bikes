Pandas does not have any built-it method to extract the week of month.
So, I am using function created by [Jan](https://stackoverflow.com/users/6089311/jan)
 
He also in-Depth explain the whole function in his answer to the [Question](https://stackoverflow.com/questions/25249033/week-of-a-month-pandas#:~:text=Now%20I%20can%20calculate%20with,calculation%20df%5B'Date'%5D 
)


```python
def weekinmonth(dates):
    """Get week number in a month.
    
    Parameters: 
        dates (pd.Series): Series of dates.
    Returns: 
        pd.Series: Week number in a month.
    """
    firstday_in_month = dates - pd.to_timedelta(dates.dt.day - 1, unit='d')
    return (dates.dt.day-1 + firstday_in_month.dt.weekday) // 7 + 1
```

```python
used_df_v1['ride_start_in_weekofmonth']=weekinmonth(used_df_v1['started_at'])
```

### Calculating the quatile 

Now we will find quatile range of the duration of ride to examine how rides are spread.

```python
quartile_of_whole_data=(used_df_v1.duration_of_ride_min.quantile([i/100 for i in range(1,101)])).to_frame().reset_index()
quartile_of_whole_data['quartile']=quartile_of_whole_data['index']*100
quartile_of_whole_data.drop(['index'], axis = 1,inplace = True)
quartile_of_whole_data=quartile_of_whole_data.reindex(columns=['quartile','duration_of_ride_min'])
quartile_of_whole_data
```

Above data confirm that 99 percent of ride are less than 131 minute.          
Now computing quatile range of data which is greater than 131 minute.

```python
duration_greater_than_131=used_df_v1[used_df_v1.duration_of_ride_min>=131]
quartile_of_100=(duration_greater_than_131.duration_of_ride_min.quantile([i/10 for i in range(1,11)])).to_frame().reset_index()
quartile_of_100['quartile']=quartile_of_100['index']*100
quartile_of_100.drop(['index'], axis = 1,inplace = True)
quartile_of_100=quartile_of_100.reindex(columns=['quartile','duration_of_ride_min'])
# quartile_of_100[0:9]
quartile_of_100
```

```python
sns.set_style('white')
sns.lineplot(data=quartile_of_100[0:9],y=quartile_of_100[0:9].duration_of_ride_min,x=quartile_of_100[0:9].quartile,color="black",linewidth=3)
plt.axvline(x=70,linestyle='--')
plt.axhline(y=241.340,linestyle='--')
plt.xlabel('Quartile')
plt.ylabel("Duration\n(min)")
plt.title("Quartile of Duration > 130")
plt.grid()
```
